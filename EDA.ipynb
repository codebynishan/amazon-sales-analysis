# removing the unnecessary column

```python
df.drop(columns=['Unnamed: 22'],inplace=True)
```

```python
df.drop(columns=['index'],inplace = True)
```

```python
df.drop(columns=['fulfilled-by'],inplace=True)
```

```python
df.drop(columns=['promotion-ids'],inplace=True)
```

# removing the row having null value 

```python
df.dropna(subset=['ship-city','ship-state','ship-postal-code','ship-country'],inplace = True)
```

# filling value in the missing column
* courier status and currency with mode
* amount with median


```python
df['Courier Status'] = df['Courier Status'].fillna(df['Courier Status'].mode()[0])
```

```python
df['currency'] = df['currency'].fillna(df['currency'].mode()[0])
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].median())
```

# converting the date column to date with format and force conversion 

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y', errors='coerce')
```

```python
df['ship-state'] = df['ship-state'].str.upper().str.strip()
```

```python
df['ship-city'] = df['ship-city'].str.upper().str.strip()
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Date'].unique
```

```python
df['Month'] = df['Date'].dt.month_name()
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df['day'] = df['Date'].dt.date
```

# value count by fulfilment 

```python
fulfillment_data = df.groupby('Fulfilment')['Status'].value_counts().unstack()
fulfillment_data
```

# cancel rate by category

```python
fulfillment_data['Cancellation Rate'] = (
    fulfillment_data['Cancelled'] / fulfillment_data.sum(axis=1)
) * 100
```

```python
cancel = df[df['Status'].str.contains('Cancel', na=False)]
total = df.groupby('Category')['Order ID'].count()
cancelled = cancel.groupby('Category')['Order ID'].count()
print((cancelled/total * 100).round(1).sort_values(ascending=False))
```

```python
ship_data = df.groupby('ship-service-level')['Status'].value_counts().unstack()
```

```python
ship_data['Cancellation Rate'] = (
    ship_data['Cancelled'] / ship_data.sum(axis=1)
) * 100
```

# cancellation rate by shipping service 

```python
plt.figure(figsize=(6,4))

bars = plt.bar(
    ship_data.index,
    ship_data['Cancellation Rate']
)

plt.title("Cancellation Rate by Ship Service Level")
plt.xlabel("Ship Service Level")
plt.ylabel("Cancellation Rate (%)")

plt.bar_label(bars, fmt='%.2f%%')

plt.show()
```

# cancellation rate by fulfillment type 

```python
plt.figure(figsize=(6,4))

bars = plt.bar(
    fulfillment_data.index,
    fulfillment_data['Cancellation Rate']
)

plt.title("Cancellation Rate by Fulfilment Type")
plt.xlabel("Fulfilment Type")
plt.ylabel("Cancellation Rate (%)")

plt.bar_label(bars, fmt='%.2f%%')

plt.show()
```

# order status 

```python
plt.figure(figsize=(12, 6))

bars = plt.bar(status_count.index, 
               status_count.values, 
               color=['#2ecc71' if 'Shipped' in x 
                      else '#e74c3c' if 'Cancelled' in x 
                      else '#f39c12' for x in status_count.index])


for i in range(len(bars)):
    plt.text(
        i,
        status_count.values[i],
        f"{status_percentage.values[i]:.1f}%\n({status_count.values[i]:,})",
        ha='center',
        va='bottom'
    )

plt.title('Order status', 
          fontsize=14, fontweight='bold')
plt.xlabel('Order Status', fontweight='bold')
plt.ylabel('Number of Orders',fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
